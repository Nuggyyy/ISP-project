```python
# Imports

import os
import random
import shutil
import numpy as np
import soundfile as sf
import librosa

SEED = 42
random.seed(SEED)
```

```python
# directory names
source_dir = "3536"
data_dir = "data"
distorted_dir = "distorted_data"

# create directory for data and distorted_data
os.makedirs(data_dir, exist_ok=True)
os.makedirs(distorted_dir, exist_ok=True)

# selected files split
files_split = {
    "normal": list(range(15)),
    "clipped": list(range(15,30)),
    "downsampled": list(range(30,45)),
    "whitenoise": list(range(45,60))
    }
```

## Step 1

Move randomly selected files to the data directory

```python
# get all flac files from source dir
directory_path = source_dir
all_files = []
for root, dirs, files in os.walk(directory_path):
    for file in files:
        if file.lower().endswith(".flac"):
            all_files.append(os.path.join(root, file))

print(f"Total flac files: {len(all_files)}")
```

```python
# randomly select 60 files and copy to data

selected_files = random.sample(all_files, 60)
for file in selected_files:
    destination = os.path.join(data_dir, os.path.basename(file))
    shutil.copy2(file, destination)

print(f"Copied {len(selected_files)} files to {data_dir}")
```

```python
print(selected_files)
```

## Step 2

Create subfolders and move audio to the designated folders

```python
# normal audio

normal_destination_path = f"{distorted_dir}/normal"
os.makedirs(normal_destination_path, exist_ok=True)

for i in files_split['normal']:
    destination = os.path.join(normal_destination_path, os.path.basename(selected_files[i]))
    shutil.copy2(selected_files[i], destination)
```

```python
# clipped audio

clipped_destination_path = f"{distorted_dir}/clipped"
os.makedirs(clipped_destination_path, exist_ok=True)

for i in files_split['clipped']:
    destination = os.path.join(clipped_destination_path, os.path.basename(selected_files[i]))
    shutil.copy2(selected_files[i], destination)
```

```python
print([selected_files[i] for i in files_split['clipped']])
```

```python
# downsampled_audio

downsampled_destination_path = f"{distorted_dir}/downsampled"
os.makedirs(downsampled_destination_path, exist_ok=True)

for i in files_split['downsampled']:
    destination = os.path.join(downsampled_destination_path, os.path.basename(selected_files[i]))
    shutil.copy2(selected_files[i], destination)
```

```python
# whitenoise

whitenoise_destination_path = f"{distorted_dir}/whitenoise"
os.makedirs(whitenoise_destination_path, exist_ok=True)

for i in files_split['whitenoise']:
    destination = os.path.join(whitenoise_destination_path, os.path.basename(selected_files[i]))
    shutil.copy2(selected_files[i], destination)
```

## Step 3

Distort the audios

```python
print(os.listdir(clipped_destination_path))
```

```python
# clipped

thresholds = [0.1,0.3,0.5,0.7,0.9]*3

for filename, threshold in zip(os.listdir(clipped_destination_path), thresholds):
    if not filename.endswith('.flac'):
        continue
    
    audio, sr = sf.read(os.path.join(clipped_destination_path, filename))
    clipped_audio = np.clip(audio, -threshold, threshold)
    sf.write(os.path.join(clipped_destination_path, filename), clipped_audio, sr)
```
