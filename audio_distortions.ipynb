```python
# Imports

import os
import random
import shutil

SEED = 42
random.seed(SEED)
```

```python
# directory names
source_dir = "3536"
data_dir = "data"
distorted_dir = "distorted_data"
```

```python
# create directory for data and distorted_data

os.makedirs(data_dir, exist_ok=True)
os.makedirs(distorted_dir, exist_ok=True)
```

```python
# get all flac files from source dir
directory_path = source_dir
all_files = []
for root, dirs, files in os.walk(directory_path):
    for file in files:
        if file.lower().endswith(".flac"):
            all_files.append(os.path.join(root, file))

print(f"Total flac files: {len(all_files)}")
```

```python
# randomly select 60 files and copy to data

selected_files = random.sample(all_files, 60)
for file in selected_files:
    destination = os.path.join(data_dir, os.path.basename(file))
    shutil.copy2(file, destination)

print(f"Copied {len(selected_files)} files to {data_dir}")
```
